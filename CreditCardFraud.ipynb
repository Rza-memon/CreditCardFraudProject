```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data = credit_card_data.dropna(how='any')
```

```python
credit_card_data.isnull().sum()
```

```python
# distribution of legit and fraud transaction
credit_card_data['Class'].value_counts()
```

```python
# 1 represents fraud transaction, 0 represents real transaction
#unbalanced data set
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
#undersampling the dataset
# Number of fraud data->56
legit_sample = legit.sample(n=60)
```

```python
new_dataset= pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting data into features and target
X = new_dataset.drop(columns='Class', axis=1)
```

```python
Y =new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2 , stratify=Y, random_state = 2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_accuracy)
```
